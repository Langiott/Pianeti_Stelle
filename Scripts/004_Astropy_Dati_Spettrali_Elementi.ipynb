# Spettri di Emissione e Assorbimento degli Elementi Chimici

Ogni elemento chimico, quando eccitato termicamente o elettricamente, emette luce a lunghezze d'onda specifiche, formando uno **spettro di righe** caratteristico. Questo principio è alla base della **spettroscopia astronomica**: analizzando la luce proveniente da stelle e galassie possiamo determinare la loro composizione chimica.

In questo notebook esploriamo le righe spettrali principali di alcuni elementi e le visualizziamo su un grafico a dispersione.

## 1. Import delle librerie

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Creazione del dataset spettrale

Definiamo le righe spettrali principali (in nm) per alcuni elementi. Ogni elemento è caratterizzato da un pattern unico di righe di emissione o assorbimento.

```python
def crea_dataset_spettrale():
    """Crea un DataFrame con le lunghezze d'onda spettrali di vari elementi."""
    data = {
        "Elemento": ["Idrogeno", "Elio", "Sodio", "Calcio", "Mercurio", "Rame", "Potassio", "Azoto"],
        "Lunghezze_d_onda_nm": [
            [121.6, 656.3, 486.1],
            [587.6, 447.1, 471.3],
            [589.0, 589.6],
            [422.7, 393.4],
            [546.1, 435.8, 404.7],
            [510.6, 578.2, 605.0],
            [766.5, 769.9],
            [337.1, 425.9]
        ]
    }
    return pd.DataFrame(data)


def esplodi_lunghezze(df):
    """Trasforma il DataFrame in formato lungo: una riga per ogni riga spettrale."""
    elementi = []
    lunghezze = []
    for _, riga in df.iterrows():
        for wl in riga['Lunghezze_d_onda_nm']:
            elementi.append(riga['Elemento'])
            lunghezze.append(wl)
    return pd.DataFrame({"Elemento": elementi, "Lunghezza_d_onda_nm": lunghezze})


df_elementi = crea_dataset_spettrale()
df_grafico = esplodi_lunghezze(df_elementi)
```

## 4. Grafico a dispersione delle righe spettrali

Ogni punto rappresenta una riga spettrale. Elementi diversi hanno colori diversi per facilitare il confronto. Le righe dell'idrogeno (Hα a 656.3 nm, Hβ a 486.1 nm) sono tra le più importanti in astronomia.

```python
def crea_grafico_spettrale(df, titolo="Lunghezze d'Onda degli Elementi Spettrali"):
    """Crea un grafico a dispersione delle righe spettrali per ogni elemento.

    Parametri:
        df: DataFrame con colonne 'Elemento' e 'Lunghezza_d_onda_nm'
        titolo: titolo del grafico
    """
    elementi_unici = df['Elemento'].unique()
    colori = plt.cm.tab10(np.linspace(0, 1, len(elementi_unici)))
    mapping_colori = dict(zip(elementi_unici, colori))

    fig, ax = plt.subplots(figsize=(12, 6))

    for elem in elementi_unici:
        subset = df[df['Elemento'] == elem]
        ax.scatter(
            subset['Lunghezza_d_onda_nm'],
            subset['Elemento'],
            color=mapping_colori[elem],
            s=100,
            edgecolors='black',
            linewidth=0.5,
            label=elem
        )

    ax.set_title(titolo, fontsize=14, fontweight='bold')
    ax.set_xlabel("Lunghezza d'Onda (nm)", fontsize=12)
    ax.set_ylabel("Elemento", fontsize=12)
    ax.grid(True, linestyle='--', alpha=0.6)
    ax.set_xlim(100, 850)

    plt.tight_layout()
    plt.show()
    return fig


crea_grafico_spettrale(df_grafico)
```

## 5. Note sull'interpretazione

- **Idrogeno**: la riga Hα a 656.3 nm (rossa) è la più intensa nel visibile; Hβ a 486.1 nm (blu-verde) e la riga Lyman-α a 121.6 nm (ultravioletto) sono fondamentali in astrofisica.
- **Elio**: scoperto prima nel Sole (1868) che sulla Terra, da cui il nome ("helios").
- **Sodio**: il doppietto a 589.0 e 589.6 nm è responsabile del colore giallo delle lampade ai vapori di sodio.
- **Calcio**: le righe H e K del Ca II a 396.8 e 393.4 nm sono tra le più forti nello spettro solare.
- **Mercurio e Rame**: usati nelle lampade a scarica per illuminazione.

La posizione e l'intensità di queste righe permettono agli astronomi di identificare la composizione chimica di stelle, nebulose e galassie lontane.
